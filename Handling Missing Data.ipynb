```
df = sqlContext.sql('SELECT * FROM ContainsNull')
```

```
df.show()
```

```
droping_nas = df.na.drop(thresh=2)
droping_nas.show()
```

```
droping_any = df.na.drop(how='any')
droping_any.show()
```

```
# drops rows with whole lines in null values
droping_how = df.na.drop(how='all')
droping_how.show()
```

```
# if any nulls in Sales' column then it drops
droping_sub = df.na.drop(subset = ['Sales'])
droping_sub.show()
```

```
# luckily this command fils only these values which are integers, its not being applied on strings
data_fill = df.na.fill(0)
data_fill.show()
```

```
# filling data only for one column
fill_column = df.na.fill('No name', subset=['Name'])
fill_column.show()
```
