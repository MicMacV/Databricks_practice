```
new_df = df.withColumn("Year", year(df['Date']))
```

```
from pyspark.sql.functions import avg
```

```
new_df.show(1)
```

```
prefinal = new_df.groupBy('Year')\
                 .agg(avg('Close').alias('Average Closing Price'))\
                 .orderBy('Year', descending=True)
```

```
prefinal.show()
```

```
result = prefinal.select(['Year', format_number('Average Closing Price', 2).alias('Average Closing Price')])
result.show()
```
